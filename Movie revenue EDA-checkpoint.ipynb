```python
# Import Libraries

import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read the movies csv file
df = pd.read_csv(r'D:\360MoveData\Users\dell\Desktop\UW\Stats coursework\movies.csv')
```

```python
df.head()
```

```python
# Find missing data

import numpy as np
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop the null values

df.dropna(inplace = True)
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types

df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.dtypes
```

```python
# Correct Years

df['year_modified'] = df['released'].astype('str').str.split(',').str[1].str[1:5]
df.head()
```

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Scatter plot with budget VS gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget VS Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color': 'red'}, line_kws = {'color': 'blue'})
```

```python
cor_mat = df.corr(method = 'pearson')
```

```python
sns.heatmap(cor_mat, annot = True)
```

```python
# There is high correlation between the movie's budget and its gross revenue
# I also want to investigate if companies filming more movies will have higher average revenue

top_count_companies = df['company'].value_counts().head(10)
top_count_companies
```

```python
# Pull out one of the top companies and get an idea of its gross earnings

df[df['company'].str.contains('TriStar')].head()
```

```python
# Plot the top 10 companies that have the highest avaerage gross earnings

company_avg_revenue = df.groupby('company')['gross'].mean().sort_values(ascending=False)

top_companies = company_avg_revenue.head(10)
top_companies.plot(kind = 'bar', color = 'skyblue')
plt.title('Top Companies by Average Gross Revenue')
plt.xlabel('Company')
plt.ylabel('Gross')
plt.xticks(rotation=45)
plt.show()
```

```python
# Find the companies that film the most with their mean gross

df[df['company'].isin(top_count_companies.index)].head()
```

```python
# Plot the 10 companies that film the most movies with their average gross revenue

top_counts = df[df['company'].isin(top_count_companies.index)]
top_counts_avg_gross = top_counts.groupby('company')['gross'].mean().sort_values(ascending = False)
top_counts_avg_gross.plot(kind = 'bar', color = 'skyblue')
plt.title('Top Film Count Companies by Average Gross Revenue')
plt.xlabel('Company')
plt.ylabel('Gross')
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy.stats import ttest_ind

log_top_avg = np.log(top_companies)
log_top_counts = np.log(top_counts_avg_gross)
t_stat, p_value = ttest_ind(log_top_avg, log_top_counts, equal_var=False)
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.6f}")

if p_value < 0.05:
    print("The means of the two lists are significantly different.")
else:
    print("The means of the two lists are not significantly different.")
```

```python
# I conclude that the companies that film the most movies do not have the highest average gross earnings.
```
